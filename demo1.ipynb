```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
#让图直接在notebook上面显示
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set()
sinplot()
```

```python
sns.set_style("whitegrid")#有格子没有背景
sinplot()
```

```python
sns.set_style("darkgrid")#灰色背景有格子
sinplot()
```

```python
sns.set_style("dark")#灰色背景没有格子
sinplot()
```

```python
sns.set_style("white")#内部没有任何背景和格子
sinplot()
sns.despine()#把上面和右边的线去掉
```

```python
sns.set_style("ticks")#有突出的小角
sinplot()
```

```python
data = np.random.random(size=(20,5))#构造数据
sns.boxplot(data=data)#画盒图
sns.set_style("dark")#设置风格
```
